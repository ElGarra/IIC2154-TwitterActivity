Antes de partir, importamos las librerías que usaremos

```python
#Código obtenido de https://www.kaggle.com/code/prathamsharma123/clean-raw-json-tweets-data
#Importing required libraries

import pandas as pd
from pandas.io.json import json_normalize
import warnings
warnings.filterwarnings("ignore")
```

Creamos el dataframe principal y el dataframe de usuarios. Ambos serán usados en las funciones que se implementarán

```python
#Código obtenido de https://www.kaggle.com/code/prathamsharma123/clean-raw-json-tweets-data

# Read JSON file containing tweets data and removce tweets not in English

raw_tweets = pd.read_json(r'farmers-protest-tweets-2021-03-5.json', lines=True)
raw_tweets = raw_tweets[raw_tweets['lang']=='en']

# Normalize 'user' field

users = json_normalize(raw_tweets['user'])
users.drop(['description', 'linkTcourl'], axis=1, inplace=True)
users.rename(columns={'id':'userId', 'url':'profileUrl'}, inplace=True)
users.head(5)

# Create DataFrame and remove duplicates

users = pd.DataFrame(users)
users.drop_duplicates(subset=['userId'], inplace=True)
```

Definimos la primera función, que retornará un top 10 de los tweets más retweeted

```python
#Función que retorna los tweets más retweeteados

def most_retweeted_tweets(df):
    df = df.nlargest(10, 'retweetCount')
    return df[['url', 'user', 'retweetCount']]

most_retweeted_tweets(raw_tweets)
```

Definimos la segunda función, que retornará el top 10 de usuarios en función de la cantidad de tweets que emitieron.

```python
#Función que retorna los usuarios más populares
def top_users(user_df):
    user_df = user_df.nlargest(10, 'statusesCount')
    return user_df[['username', 'userId', 'statusesCount']]

top_users(users)
```

Por último, implementamos la función que retorna el top 10 de días en los que han habido más tweets.

```python
#Función que retorna los días en que han habido más tweets
def top_days(df):
    df = df.groupby(df['date'].dt.date).size().reset_index(name='Counts')
    df = df.nlargest(10, 'Counts')
    return df

top_days(raw_tweets)
```

```python
def main(df, user_df):
    most_retweeted_tweets(df)
    top_users(user_df)
    top_days(df)

if "__name__" == "__main__":
    main()
```
